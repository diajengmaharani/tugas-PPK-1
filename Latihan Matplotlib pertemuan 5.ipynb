### Latihan Matplotlib
1. Cari bagaimana cara menampilkan judul(title) dari grafik
2. Cari bagaimana cara memberikan label pada sumbu x dan sumbu y
3. Cari bagaimana cara menambahkan teks dalam grafik

Misal diberikan sebuah nilai x dan y berikut:

```python
# import library
import matplotlib.pyplot as plt  

#persiapan data
x = [0,1,2,3,4,5] 
y= [0,5,10,15,20,25] 

#1 memberikan judul pada grafik
plt.title("Grafik Linear") 
  
#membuat plot grafik
plt.plot(x, y, color='green')  

#2. memberikan label pada sumbu x dan ya
plt.xlabel('sumbu x')  
plt.ylabel('sumbu y') 

#3 menambahkan teks pada grafik
plt.text(4,16, 'Grafik titik antara x dan y', fontsize=12, color='g')

# menyimpan grafik
plt.savefig('foo.png')

#menampilkan grafik
plt.show()
```
